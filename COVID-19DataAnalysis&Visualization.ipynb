## Step 2: Data Preparation

### Handle Missing Values in the DataFrame

```python
# Drop the columns with over 200 missing values
df = df.drop(columns=['NewCases', 'NewDeaths', 'NewRecovered'])
df
```

```python
# Replace the missing values in 'Serious,Critical' & 'Population' & 'TotalDeaths' & 'TotalRecovered' & 
#'ActiveCases' & 'Tot Cases/1M pop' & 'Deaths/1M pop' & 'TotalTests' & 'Tests/1M pop' columns with their mean values
df['Serious,Critical'] = df['Serious,Critical'].fillna(value=df['Serious,Critical'].mean())
df['Population'] = df['Population'].fillna(value=df['Population'].mean())
df['TotalDeaths'] = df['TotalDeaths'].fillna(value=df['TotalDeaths'].mean())
df['TotalRecovered'] = df['TotalRecovered'].fillna(value=df['TotalRecovered'].mean())
df['ActiveCases'] = df['ActiveCases'].fillna(value=df['ActiveCases'].mean())
df['Tot Cases/1M pop'] = df['Tot Cases/1M pop'].fillna(value=df['Tot Cases/1M pop'].mean())
df['Deaths/1M pop'] = df['Deaths/1M pop'].fillna(value=df['Deaths/1M pop'].mean())
df['TotalTests'] = df['TotalTests'].fillna(value=df['TotalTests'].mean())
df['Tests/1M pop'] = df['Tests/1M pop'].fillna(value=df['Tests/1M pop'].mean())
```

```python
# Drop the rows that contain missing values
df = df.dropna()
df
```

```python
# Confirm that the dataset has no more missing values
df.info()
```

### 3. Top 3 Continents with the highest deaths per 1M population

```python
# Analysis
Subset = df.dropna(subset=['Deaths/1M pop'])
GroupedData = Subset.groupby('Continent')['Deaths/1M pop'].sum()

# Sorting the names in the descending order
SortedData = GroupedData.sort_values(ascending=False).head(3)
print(SortedData)
```

```python
# Visualization
plt.figure(figsize=(6.5, 5));
sns.barplot(x=SortedData.values, y=SortedData.index, alpha=0.8);
plt.xlabel('Number of Deaths per 1M Population');
plt.ylabel('Continent');
plt.title('Top 3 Continents with the highest deaths per 1M population', fontsize=18);
```

### 4. Top 3 Continents with the highest Serious,Critical patients

```python
# Analysis
Subset = df.dropna(subset=['Serious,Critical'])
GroupedData = Subset.groupby('Continent')['Serious,Critical'].sum()

# Sorting the names in the descending order
SortedData = GroupedData.sort_values(ascending=False).head(3)
print(SortedData)
```

```python
# Visualization
plt.figure(figsize=(6.5, 5));
custom_colors = ['#FF5733', '#3498DB', '#27AE60']
sns.barplot(x=SortedData.index, y=SortedData.values, alpha=0.8, palette=custom_colors);
plt.xlabel('Continent');
plt.ylabel('Number of Serious,Critical Patients');
plt.title('Top 3 Continents with the highest Serious,Critical patients', fontsize=18);
plt.xticks(rotation=45);
```
